```python
import pandas as pd
```

```python
gtf = pd.read_csv('/Volumes/Samsung_T5/ncVEP/VCFs/gencode.v46.chr_patch_hapl_scaff.annotation.gtf', sep='\t', comment='#', header=None)
gtf.columns = ['chrom', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']
```

```python
exons = gtf[gtf['feature'] == 'exon']
exons = exons[exons['chrom'].str.startswith('chr')]
exons
```

```python
#save exons to bed file '/Volumes/Samsung_T5/ncVEP/VCFs/gencode_exons.bed'
exon_bed_out = exons[['chrom', 'start', 'end']]
exon_bed_out.to_csv('/Volumes/Samsung_T5/ncVEP/VCFs/gencode_exons.bed', sep='\t', index=False, header=False)
```
